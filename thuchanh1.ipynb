Nhập vào năm bất kỳ nếu có dữ liệu thì thực hiện các việc sau:  
1. Vẽ biểu đồ giá vàng của năm đó với màu xanh
2. vẽ biểu đồ giá vàng của năm trước đó với vàng
3. Hiền thị giá vàng nhỏ nhất của năm đó
4. Hiền thị giá vàng lấn nhất của năm đó
5. Hiền thị giá vàng trung bình của năm đó
6. Hiền thị độ lệch chuẩn của giá vàng năm đó
7. Hiển thị 2 tháng có giá vàng nhỏ nhất của năm đó
8. Vẽ biểu đồ giá vàng trên toàn bộ dữ liệu.

```python
#  Tải dữ liệu dùng thư viện pandas
import pandas as pd
gia_vang=pd.read_csv('gold_price.csv')
# thông tin về dữ liệu đã đọc được
gia_vang.info()
```

```python
# Chuyển cột "Date" sang kiểu dữ liệu ngày tháng
gia_vang.Date=pd.to_datetime(gia_vang['Date'])
# đọc lại thông tin
gia_vang.info()
```

```python
# Câu 8: Vẽ biểu đồ giá vàng trên toàn dữ liệu
import matplotlib.pyplot as plt
#x=gia_vang.Date
#y=gia_vang.Price
#plt.plot(x,y)
plt.plot(gia_vang.Date,gia_vang.Price)
```

```python
# Câu 1 + 2
# NHập năm cần xem
year_in=input() # string chưa chuỗi ký tự nhập từ bàn phím
numpy_gia_vang=gia_vang.to_numpy()
x=[]
y=[]
z=[]
for ngaythang,gia in numpy_gia_vang:    
    if ngaythang.year==int(year_in) : # Nếu năm của dòng hiện tại == năm nhập vào thì lưu dữ liêu  
        x.append(ngaythang) # lưu date vào biến x
        y.append(gia) # lưu giá vàng của date hiện tại vào biến y
    elif ngaythang.year==int(year_in)-1:
        z.append(gia)
# chúng ta thu được dữ liệu của 12 tháng trong năm nhập vào
# vẽ biểu đồ năm nhập vào
plt.plot(x,y,'b-',label=year_in)
# biểu đồ năm trước năm nhập vào
plt.plot(x,z,'y-',label=str(int(year_in)-1))
plt.xlabel("Date")
plt.ylabel("Giá vàng USD")
plt.title("Biểu đồ giá vàng năm " + year_in)
plt.legend()

```

```python
print('Giá vàng thấp nhất năm ' + year_in + ' là', min(y),'USD')
```

```python
# câu 7
import numpy as np
numpy_gia=np.array(y) # Ép kiểu biến y từ dạng list sang array-numpy
tmp=numpy_gia.argsort()
print('2 tháng có giá vàng thấp nhât trong năm',year_in,'là ',
    tmp[0]+1,'và',tmp[1]+1)
```
