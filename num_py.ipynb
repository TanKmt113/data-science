```python
import numpy as np
import math
x=np.random.rand(10,2)
print(x)
radian_360=np.arange(0,math.pi*2,0.05)
print(radian_360)
```

** Các biến sử dụng trong chương trình

```python
sinx=np.sin(radian_360)
print(sinx)
```

* Đoạn chương trình sau thí nghiệm hàm plot trong thư viện matplotlib để vẽ đường hinh sin cho các điểm (x,y)

```python
import matplotlib.pyplot  as plt # khai báo thư viện matplotlib
plt.plot(radian_360,sinx) # vẽ đường hình sin mô tả các điểm ảnh (x,y)
```

```python
plt.plot(radian_360,sinx) # vẽ biểu đồ hình sin với dữ liệu các điểm (x,y)=(radian_360,sin(radina_360))
plt.xlabel('x in radian') # gán nhãn trục hoành
plt.ylabel('sin(x)') # gán nhãn trục tung
plt.title('Test plot of (x,sin(x))') # gán nhã cho đồ thị
plt.show() # hiển thị các nội dung đã gán nhãn ở trên lên biểu đồ
```

Câu hỏi: 3 bạn có điểm thấp nhất là ai?
Hàm "phần đoạn" trong numpy
Cú pháp:
y=np.patition(x,k)
trong đó: 
+ x là mảng cần phận đoạn
+ k là số phần tử nhỏ nhất cần trả về
+ Kết mảng x bị phân tách thành 2 nửa. Trong đó nửa trái là k phần tử nhỏ nhất, nửa phải là phần còn lại được sắp xếp theo chiều tang dần.


```python
x=np.random.rand(10)
print('x=',x)
y=np.partition(x,3)
print('y=',y)
```
