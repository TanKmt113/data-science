```python
import datetime as dt
import matplotlib.pyplot as plt
import numpy as np
fig=plt.figure()
ax=plt.axes()

inyear=input()
x=[]
y=[]
z=[]

for row in np_price:    
    date,pri=row        
    if date.year==int(inyear):
        x.append(date)
        y.append(pri)
    elif date.year==int(inyear)-1:
        z.append(pri)

ax.plot(x,y,'b-',label=inyear)
ax.plot(x,z,'r-',label=str(int(inyear)-1))
ax.set_xlabel('Month')
ax.set_ylabel("USD")
ax.legend()

```
