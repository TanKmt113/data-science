***Vẽ biểu đồ trong hệ 3D***  
**Các bước:**
* Bước 1: Khai báo hệ tọa độ 3D



```python
he3d=plt.axes(projection='3d') # Thực hiện khai báo 1 hệ tọa độ 3D
# VD vẽ đường trong không gian 3D
zline=np.linspace(0,15,1000) # Khỏi tạo 1000 con số trong khoảng từ 0-15
xline=np.sin(zline)
yline=np.cos(zline)
#print(zline.shape,x)
# Vẽ biểu đồ
he3d.plot3D(xline,yline,zline)
```

* Bước 2: Dựng dữ liệu cho 3 trục x,y,z
* Bước 3: Vẽ dùng hàm plot3D(x,y,z)

```python
# Định nghĩa hàm z=f(x,y), hàm này chỉ có tính chất ví dụ để sinh dữ liệu
# Trong thực tế các dữ liệu 3 trục x,y,z là có sẵn, không cần sinh tự động
def f(x,y):
    return np.sin(np.sqrt(x**2 + y**2))  
# Sinh ngẫu nhiễn dữ liệu trục x và y
#x= np.linspace(-6,6,30)
x=np.random.randint(0,100,50)
y=np.random.randint(200,250,50)
#print(x)
#y= np.linspace(-6,6,30)
# Tạo ra các lưới trong không gian đặc
X,Y = np.meshgrid(x,y)
# tính giá trị của Z trong mọi điểm của không gian 3D
Z=f(X,Y)

# vẽ dùng contour3D
he3d2=plt.axes(projection='3d')
he3d2.contour3D(X,Y,Z,10)
#plt.show()
```
