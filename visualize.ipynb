```python
import numpy as np 
import matplotlib.pyplot as plt
x = np.linspace(0, 5, 50)
#y = np.linspace(0, 5, 50)
y=np.random.rand(50)
#z = np.arange(50)
z=np.random.randint(1,50,50)
color = np.arange(50)
plt.scatter(x,y,s=z,c=color,marker='o')
y2=np.random.rand(50)
plt.scatter(x,y2,s=z,c=color,marker='^')
```
